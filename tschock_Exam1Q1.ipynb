```python
# Import necessary libraries
import csv
import random
import math
import copy
import pandas as pd
import numpy as np
```

```python
def create_teams(team_size, output_file):
    # Create dictionaries to store student name and email addresses together
    name_email = pd.read_csv('./DataCamp-members-20240913174631.csv', delimiter=',', names=['email', 'name', 'blank', 'role', 'learn_license', 'workspace_license'], header=0, index_col=False)
    # Remove unnecessary columns 
    name_email = name_email.drop(['blank', 'role', 'learn_license', 'workspace_license'], axis=1)
    # Remove any excess white space
    name_email['email'] = name_email['email'].str.strip()
    
    # Assign name and email values to shuffled_students variable
    shuffled_students = name_email.values.tolist()
    # Randomly shuffle the students
    random.shuffle(shuffled_students)
    
    # next, parse the list into teams of size team_size
    num_teams = math.ceil(len(name_email)/team_size)

    # Most of this code chunk comes from chatGPT, as I kept getting duplicate student names and emails in my teams output file, so I asked it to "rewrite my code to fix the issue of duplicate student names and emails in the output file"
    # chatGPT also incorporates further error checking as it allows the user to change the team_size variable to any number and the output file will work as it should, given the number is in bounds
    # Create teams
    teams = []
    assigned_students = set()
    for i in range(num_teams):
        team = []
        while len(team) < team_size and shuffled_students:
            student = shuffled_students.pop(0)
            if student[1] not in assigned_students:
                team.append(student)
                assigned_students.add(student[1])
        team_emails = [student[0] for student in team] 
        print("Team Emails:", team_emails)
        team_names = [student[1] for student in team] 
        print("Team Names:", team_names)
        team_info = []
        for student in team:
            team_info.append(student[1] + ' // ' + student[0])  # Combine name and email
        # Pad the team_info list with empty strings if necessary
        while len(team_info) < team_size:
            team_info.append('')
        teams.append(team_info)

    # Create a DataFrame from the teams
    data = []
    for team in teams:
        # Learned the {} from chatGPT - updates incrementally using index + 1
        team_data = ['Team {}'.format(teams.index(team) + 1)] 
        # Extends all elements of team to the end of team_data
        team_data.extend(team)
        # Adds team_data back into the empty data list
        data.append(team_data)

    # I asked chatGPT to create a header row for each student number - {}
    teams_df = pd.DataFrame(data, columns=['Team'] + ['Student {} Name and Email'.format(i+1) for i in range(team_size)])

    # Write the DataFrame to the output file
    teams_df.to_csv(output_file, index=False)
```

```python
create_teams(3, 'Student_Teams.csv')
```
